#### Writing Results

```python
import pickle
```

```python
def writeResults(obj):
  file_path = f'/content/gdrive/My Drive/GWO/Results/{obj["dataset"]}_{obj["iterations"]}.txt'
  resultFile = open(file_path, 'wb')
  pickle.dump(obj, resultFile)
  resultFile.close()
```

###### Iterations

```python
iterations = 100
```

#### Test all datasets

```python
for dataset in datasets:

  print("\n\nDataset : ", dataset['name'])
  dataframe = pd.read_csv( dataset_dir_path + dataset['path'] )
  bgwo = BMOGWO(dataset = dataframe , greyWolvesNum=8, archiveSize=50, maxIt= iterations)

  start_time = time.time()
  bgwo.optimize()
  end_time = time.time()

  result = {
      'dataset': dataset['name'],
      'dim': bgwo.dim,
      'iterations': bgwo.maxIt,
      'archive': [sol.position for sol in bgwo.archive],
      'archiveCosts': [sol.cost for sol in bgwo.archive],
      'explored': bgwo.explored,
      'runtime': end_time - start_time
  }

  writeResults(result)
```
